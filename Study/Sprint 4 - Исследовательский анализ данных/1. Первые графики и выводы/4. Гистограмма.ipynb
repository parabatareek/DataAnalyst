# 1.
Медианные и средние значения недостаточно характеризуют данные. Настало время посмотреть на распределение значений.
Постройте гистограмму по значениям времени, проведённого на АЗС. Эти значения хранятся в столбце time_spent.

Подсказка:
Примените к датафрейму метод hist(), передав в качестве аргумента столбец time_spent.

```python
import pandas as pd

data = pd.read_csv ('../datasets/visits.csv', sep='\t')
data['time_spent'].hist()
```

# 2.
Предыдущая гистограмма выглядит странно, потому что максимальное значение столбца time_spent сильно превышает большинство других значений. Измените код, чтобы гистограмма стала более информативной.
Постройте новую гистограмму и увеличьте число корзин до 100.

Подсказка:
Передайте в bins= число корзин, равное 100.

```python
data['time_spent'].hist(bins=100)
```

3.
Итак, гистограмма стала более информативной. Можно исключить слишком большие значения времени заправки и посмотреть на остальные.
Используйте параметр range, чтобы изучить распределение значений time_spent, находящихся в диапазоне от 0 до 1500. Количество корзин оставьте прежним — 100.

Подсказка:
Передайте аргументу range наименьшее и наибольшее значения в круглых скобках.

```python
data['time_spent'].hist(bins=100, range=(0, 1500))
```
