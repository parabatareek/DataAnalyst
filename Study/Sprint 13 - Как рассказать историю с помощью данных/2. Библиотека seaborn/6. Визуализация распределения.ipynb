Постройте график совместного распределения данных таблицы rest_data.

Подсказка
Постройте график совместного распределения методом sns.pairplot().

```python
import seaborn as sns
import pandas as pd

rest_data = pd.read_csv('datasets/restaurant_data.csv')
rest_data['bill_datetime'] = pd.to_datetime(rest_data['bill_datetime'])
sns.pairplot(data=rest_data)
```

Добавьте третье измерение — пол официанта 'waiter_sex'.

Подсказка
Передайте столбец с полом официанта аргументу hue метода sns.pairplot().

```python
import seaborn as sns
import pandas as pd

rest_data = pd.read_csv('datasets/restaurant_data.csv')
rest_data['bill_datetime'] = pd.to_datetime(rest_data['bill_datetime'])

sns.pairplot(data=rest_data,
             hue='waiter_sex')
```
