Укажите цвет '#006400' для графика из прошлого урока.

Подсказка
Параметру color метода jointplot() передайте строку с HEX-кодом из условия.

```python
import seaborn as sns
import pandas as pd

rest_data = pd.read_csv('datasets/restaurant_data.csv')
rest_data['bill_datetime'] = pd.to_datetime(rest_data['bill_datetime'])
sns.jointplot(x='tips', y='restaurant_bill', data=rest_data, kind='reg', color='#006400')
```
