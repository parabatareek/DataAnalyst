Измените стиль графика на 'dark'.

Подсказка
Передайте методу sns.set_style стиль 'dark'.

```python
import seaborn as sns
import pandas as pd

rest_data = pd.read_csv('datasets/restaurant_data.csv')
rest_data['bill_datetime'] = pd.to_datetime(rest_data['bill_datetime'])

sns.jointplot(
    x='tips', y='restaurant_bill', data=rest_data, kind='reg', color='#006400'
)
sns.set_style('dark')
```
