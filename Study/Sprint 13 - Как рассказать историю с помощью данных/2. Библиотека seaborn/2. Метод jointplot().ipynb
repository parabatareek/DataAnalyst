# 3.
Постройте график зависимости общего счёта от чаевых.

Подсказка
Передайте методу jointplot() чаевые по оси X, общий счёт — по оси Y. Укажите данные, по которым строить график и передайте 'reg' параметру kind.

```python
import seaborn as sns
import pandas as pd

rest_data = pd.read_csv('datasets/restaurant_data.csv')
rest_data['bill_datetime'] = pd.to_datetime(rest_data['bill_datetime'])
sns.jointplot(data=rest_data,
              x='tips',
              y='restaurant_bill',
              kind='reg')
```
