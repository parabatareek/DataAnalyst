Вы работаете аналитиком сети ресторанов. Есть основной ресторан и 4 филиала. Из них №3 — самый прибыльный по результатам месяца. Это вызвало сомнения у руководства: ресторан находится в малопроходимом районе. Изучите названия столбцов датафрейма. Постройте столбчатую диаграмму числа заказов в каждом ресторане. Используйте стиль 'seaborn-talk'.
Путь к файлу: /datasets/rest.csv

Подсказка
Постройте столбчатую диаграмму так: plt.bar(rest['name_rest'],rest['orders']). Вызовите метод plt.show() для вывода графика.

```python
import matplotlib.pyplot as plt
import pandas as pd

rest = pd.read_csv(filepath_or_buffer='datasets/rest.csv')# прочитайте файл и сохраните в переменной
plt.bar(x=rest['name_rest'],
        height=rest['orders'])
plt.style.use('seaborn-talk')
plt.show()
print(rest)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

rest = pd.read_csv(filepath_or_buffer='datasets/rest.csv')
plt.bar(x=rest['name_rest'],
        height=rest['avg_bill'])
plt.style.use('seaborn')
plt.show()
```
