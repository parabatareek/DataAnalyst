# 1.
Оцените значение функции ошибки для подмножества из нашего примера. Пусть вы строите модель для прогнозирования величины, которая принимает значения:
y = [28.1, 18.7, 1.0, 10.2, 11.6, 19.9, 24.4, 18.1, 18.5, 25.0, 21.8, 13.4, 18.0, 11.1, 21.1].
У вас есть только один признак, который принимает значения:
x = [12, 9, 4, 5, 5, 7, 7, 6, 7, 11, 9, 8, 10, 4, 11].
При этом вы оцениваете, насколько хорошо ваши данные описывают функции y = 2 + 2*x и y = 3 + 1*x.
Постройте прогнозы двумя уравнениями и выведите их.

Подсказка:
Примените генераторы списков для расчёта векторов y_1 и y_2 по формулам двух функций из условия.

```python
import numpy as np

# Создайте векторы x и y
x = [12, 9, 4, 5, 5, 7, 7, 6, 7, 11, 9, 8, 10, 4, 11]# ваш код здесь
y = [28.1, 18.7, 1.0, 10.2, 11.6, 19.9, 24.4, 18.1, 18.5, 25.0, 21.8, 13.4, 18.0, 11.1, 21.1]# ваш код здесь

y_1 = [(2 + 2*i) for i in x] # рассчитайте прогноз первой функцией
y_2 = [(3 + 1*i) for i in x] # рассчитайте прогноз второй функцией
print('Прогноз первой моделью:', y_1)
print('Прогноз второй моделью:', y_2)
```

# 2.
Создайте функцию расчёта среднеквадратичной ошибки по формуле из конца урока.
Рассчитайте значения функции ошибки для двух прогнозов.
Выведите их на экран. Какая модель лучше?

Подсказка:
После перевода списков в формат np.array с ними можно делать поэлементные операции, как с обычными числами — и вычислять их сумму.

```python
import numpy as np

# Создайте векторы x и y
x = [12, 9, 4, 5, 5, 7, 7, 6, 7, 11, 9, 8, 10, 4, 11]# ваш код здесь
y = [28.1, 18.7, 1.0, 10.2, 11.6, 19.9, 24.4, 18.1, 18.5, 25.0, 21.8, 13.4, 18.0, 11.1, 21.1]# ваш код здесь

y_1 = [2 + 2*i for i in x] # раcсчитайте прогноз первой функцией
y_2 = [3 + 1*i for i in x] # раcсчитайте прогноз второй функцией

def error_function(y_real, y_pred):
    q =  sum((np.array(y_real) - np.array(y_pred))**2) / len(y_real)  # Ваш код здесь
    return q

q_1 = error_function(y_real=y, y_pred=y_1) # рассчитайте ошибку для прогноза первой модели
q_2 = error_function(y_real=y, y_pred=y_2) # рассчитайте ошибку для прогноза второй модели

print('Ошибка первой модели:', q_1)
print('Ошибка второй модели:', q_2)
```
