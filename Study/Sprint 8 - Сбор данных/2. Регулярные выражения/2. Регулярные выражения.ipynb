# 2.
Напишите get-запрос к странице https://store.data-analyst.praktikum-services.ru/. Сохраните текстовое содержание запроса в переменной req_text. Напишите регулярное выражение, которое получает информацию внутри тегов <title> </title>. Выведите результат в следующем виде : <title>Текст внутри тега</title>.

Подсказка
Получите текстовое содержание запроса: примените метод get() библиотеки requests с атрибутом text. Параметром методу get() передайте переменную URL.
Информация между <title> и </title> содержит русские и английские буквы и пробел. Одно и более вхождений обозначают шаблоном +. Напишите шаблон регулярного выражения, окружённый тегами <title> и </title>.

```python
import requests
import re

URL = 'https://store.data-analyst.praktikum-services.ru/'# сохраните ссылку на сайт
req_text = requests.get(url=URL).text# получите текстовое содержание запроса
print(re.search("<title>[A-zА-я ]+</title>", req_text).group())
```

# 3.
Напишите регулярное выражение, которое выводит все названия продуктов, начинающихся со слова Масло. Не включайте в название жирность и массу продукта.

Подсказка
Все названия продуктов указаны на русском и английском языке, могут содержать пробелы и дефисы.


```python
import requests
import re

URL = 'https://store.data-analyst.praktikum-services.ru/'
req_text = requests.get(URL).text
print(re.findall('Масло [A-zА-я -]+', req_text))
```

# 4.
Напишите регулярное выражение, которое найдет все названия, начинающиеся со слова Молоко. Не включайте в название жирность и массу продукта. Сохраните результат в переменной found_products.
Выведите сначала количество таких продуктов, а на следующей строке — сами продукты.

Подсказка
Методом findall() найдите все названия продуктов, начинающиеся со слова молоко. Наименования могут содержать русские и латинские буквы, а также пробелы и дефисы.
Укажите, что поиск соответствующих шаблону подстрок нужно вести в строке req_text.
Функцией len() найдите количество продуктов, удовлетворяющих условию.

```python
import requests
import re

URL = 'https://store.data-analyst.praktikum-services.ru/'
req_text = requests.get(URL).text
found_products = re.findall(pattern='Молоко [A-zА-я -]+',
                            string=req_text)
print(len(found_products))
print(found_products)
```
