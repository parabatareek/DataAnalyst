# 1.
Получите прогноз погоды на 5 дней для региона Moscow.

Подсказка
В словаре параметров запросов передайте ключу "limit" значение 5.

```python
import requests

BASE_URL = "https://weather.data-analyst.praktikum-services.ru/v1/forecast"
# URL метода

params = { # словарь с параметрами запроса
    "city" : "Moscow",
    'limit' : 5
}

response = requests.get(BASE_URL, params=params)
print(response.text)
```

Получите прогноз погоды на 5 дней с детальной информацией об осадках для региона Moscow.
Однако без прогноза погоды по часам — передайте этому ключу словаря значение False.
Выведите на экран текстовое содержание запроса.

Подсказка
В словаре параметров запросов params укажите ключи: "city", "limit", "hours", "extra". Присвойте им значения: "Moscow", 5, False и True соответственно.
Передайте функции print() конструкцию response.text.

```python
import requests

BASE_URL = "https://weather.data-analyst.praktikum-services.ru/v1/forecast"
# URL метода

params = {'city':'Moscow',
          'hours':False,
          'limit':5,
          'extra':True}

response = requests.get(BASE_URL, params=params)

#ваш код здесь
print(response.text)
```
