```python
query = '''SELECT b.title, COUNT(re.review_id) as review_count, AVG(ra.rating) as rating_avg
           FROM books b LEFT JOIN reviews re ON (b.book_id = re.book_id)
                LEFT JOIN ratings ra ON (b.book_id = ra.book_id)
           GROUP BY b.book_id, b.title
           ORDER BY rating_avg DESC
'''
data = pd.io.sql.read_sql(query, con=engine)
display(data)
```

## –ó–∞–¥–∞–Ω–∏–µ 3: –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ –∏–∑–¥–∞—Ç–µ–ª—å—Å—Ç–≤–æ, –∫–æ—Ç–æ—Ä–æ–µ –≤—ã–ø—É—Å—Ç–∏–ª–æ –Ω–∞–∏–±–æ–ª—å—à–µ–µ —á–∏—Å–ª–æ –∫–Ω–∏–≥ —Ç–æ–ª—â–µ 50 —Å—Ç—Ä–∞–Ω–∏—Ü ##

```python
query = '''SELECT p.publisher, COUNT(b.book_id) as books_count
           FROM books b INNER JOIN publishers p on (b.publisher_id = p.publisher_id)
           WHERE b.num_pages > 50
           GROUP BY p.publisher
           ORDER BY books_count DESC
           LIMIT 1
'''
data = pd.io.sql.read_sql(query, con=engine)
display(data)
```

## –ó–∞–¥–∞–Ω–∏–µ 4:
1. –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ –∞–≤—Ç–æ—Ä–∞ —Å —Å–∞–º–æ–π –≤—ã—Å–æ–∫–æ–π —Å—Ä–µ–¥–Ω–µ–π –æ—Ü–µ–Ω–∫–æ–π –∫–Ω–∏–≥;
2. —É—á–∏—Ç—ã–≤–∞–π—Ç–µ —Ç–æ–ª—å–∫–æ –∫–Ω–∏–≥–∏ —Å 50 –∏ –±–æ–ª–µ–µ –æ—Ü–µ–Ω–∫–∞–º–∏;

```python
query = '''SELECT a.author,
                  COUNT(ra.rating_id) as rating_count,
                  AVG(ra.rating) as rating_avg
           FROM books b INNER JOIN ratings ra ON (b.book_id = ra.book_id)
                INNER JOIN authors a ON(b.author_id = a.author_id)
           GROUP BY a.author
           HAVING COUNT(ra.rating_id) > 50
           ORDER BY rating_avg DESC
           LIMIT 1'''
data = pd.io.sql.read_sql(query, con=engine)
display(data)
```

```python
query = '''
WITH
rating_count AS (SELECT ra.username
                 FROM ratings ra
                 GROUP BY ra.username
                 HAVING COUNT(ra.rating) > 50),
reviews_count AS (SELECT re.username, COUNT(re.review_id) reviews_count
                  FROM reviews re
                  GROUP BY re.username)
SELECT AVG(rg.reviews_count) reviews_avg
FROM rating_count rc LEFT OUTER JOIN reviews_count rg ON (rc.username = rg.username)
ORDER BY reviews_avg
'''
data = pd.io.sql.read_sql(query, con=engine)
data
print(f'–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±–∑–æ—Ä–æ–≤ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –ø–æ—Å—Ç–∞–≤–∏–≤—à–∏—Ö –±–æ–ª–µ–µ 50 –æ—Ü–µ–Ω–æ–∫: {data["reviews_avg"][0]}')
```
